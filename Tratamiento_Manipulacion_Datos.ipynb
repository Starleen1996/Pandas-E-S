#### Aplicacos en normalize para desanidar diccionarios de la columna "dados_ventas"

```python
datos_ventas_normalizados = pd.json_normalize(datos_ventas['dados_vendas'])
```

```python
datos_ventas_normalizados
```

```python
datos_ventas = datos_ventas_normalizados
```

```python
# Colectar los valores de las columnas y verificar
columnas = list(datos_ventas.columns)
columnas
# Destrincar las listas con explode
datos_ventas = datos.explode(columnas[1:])
# Resetear los index de las líneas
datos_ventas.reset_index(drop=True,inplace=True)
# Observar el DataFrame
datos

# Verificar los tipos de datos con info
datos_ventas.info()

# La columna numérica es el 'Valor da compra'
datos_ventas['Valor da compra']

# Iniciar la transformación
# Import de la biblioteca numpy
import numpy as np
# Remover los textos presentes en la base
# Cambiar las comas separadoras del decimal por punto
datos_ventas['Valor da compra'] = datos_ventas['Valor da compra'].apply(lambda x: x.replace('R$ ', '').replace(',','.').strip())
# Cambiar los tipo de datos para float
datos_ventas['Valor da compra'] = datos_ventas['Valor da compra'].astype(np.float64)
# Verificar la transformación
datos_ventas.info()
```

## Datos Numericos

```python
columnas = list(administracion_condominios_normalizados.columns)
```

```python
columnas
```

```python
# El método explode() "desenrolla" (o "desanida") listas contenidas en columnas.
# columnas[1:]:Esto selecciona todas las columnas excepto la primera de la lista columnas. Por ejemplo, si columnas = ['ID', 'Servicios', 'Pagos', 'Contactos'], entonces columnas[1:] será ['Servicios', 'Pagos', 'Contactos'].
datos = administracion_condominios_normalizados.explode(columnas[1:])
datos
```

```python
# Reseteando los indices
#.reset_index(): Restaura el índice del DataFrame al valor predeterminado (es decir, un contador incremental desde 0).

# drop=True: Le dice a pandas que no conserve la columna del índice anterior como una columna nueva en el DataFrame. Si lo dejaras como drop=False (valor por defecto), pandas movería el índice viejo a una nueva columna.

#inplace=True: Modifica el DataFrame en el lugar, sin necesidad de reasignarlo a una variable nueva.


datos.reset_index(drop=True, inplace=True)
datos.head()
```

```python
datos.info()
```

```python
import numpy as np
```

```python
# Esta linea limpia todos los signos de $,'', y espacios de la columna valor_aluguel
#Nota: el apply solo funciona para convertir 1 sola columna es decir una serie.
#nOTA: Cuando se trabajan con dos columnas ya estamos hablando de un dataframe, por ende no se aplica el apply, sino el "applymap"
datos['valor_aluguel']=  datos['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','').replace('reais','').strip())
```

```python
datos.info()
```

```python
datos.head()
```

```python
# .astype(np.float64): Esta función convierte los datos de esa columna a tipo float64 de NumPy, es decir, números decimales de doble precisión.
datos['valor_aluguel'] =datos['valor_aluguel'].astype(np.float64)
```

```python
datos.info()
```

## DATOS DE TEXTO

```python
datos_ventas = pd.read_json('dados_vendas_clientes.json')
```

```python
datos_ventas
```

```python
datos_ventas_normalizados = pd.json_normalize(datos_ventas['dados_vendas'])
```

```python
datos_ventas_normalizados
```

```python
columnas = list(datos_ventas_normalizados.columns)
```

```python
datos_2= datos_ventas_normalizados.explode(columnas[1:])
```

```python
datos_2
```

```python
datos_2.reset_index(drop=True, inplace=True)
datos_2
```

```python
datos_2['Cliente'] =datos_2['Cliente'].str.lower() # convierte la seria a string y lo coloca en minuscula con el metodo lower()
```

```python
datos_2.head()
```

```python
datos_2['Cliente'][0]
```

```python
# Utilizando el metodo Regex
# Si quieres eliminar todo lo que NO sea texto (letras) o números, necesitas una negación en una expresión regular que conserve solo: [^a-zA-Z0-9]
datos_2['Cliente'] =datos_2['Cliente'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
datos_2
```

```python
# Tokenización de string
# separar la cadena de texto en elementos
#.str.split() es un método de pandas que se usa para dividir cadenas de texto dentro de una columna.
datos_2['Cliente'] =datos_2['Cliente'].str.split('')
```
