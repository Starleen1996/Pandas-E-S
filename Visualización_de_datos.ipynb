## Visualización de datos con Python: Matplotlib,Seaborn y Plotly

```python
import pandas as pd
```

```python
datos = pd.read_csv('/content/inmigrantes_canada.csv')
print(datos.head(5))
```

```python
print(datos.columns)
```

```python
datos.columns = datos.columns.str.strip()
```

```python
print(datos.columns.tolist())
```

```python
datos.set_index('Pais', inplace=True)
datos
```

```python
#Crea una lista de strings (cadenas de texto) que van desde el año 1980 hasta el 2013 (el range en Python excluye el último número, por eso no incluye el 2014).
anos = list(map(str, range(1980, 2014)))
anos
```

```python
#.loc[] es una forma de acceder a datos por etiquetas en pandas.
# Se refiere a la etiqueta de la fila. Para que esto funcione, el índice (index) del DataFrame datos debe tener la columna 'Pais' configurada como índice, lo cual ya hiciste antes con:
colombia = datos.loc['Colombia', anos]
colombia
```

```python
#Ese fragmento de código convierte los datos de inmigración de Colombia en un diccionario Python que será útil, por ejemplo, para crear un DataFrame o visualizar los datos. Vamos paso a paso:
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4)) # Con est código se establece el tamaño de la figura
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])  #Grafico Años en el eje X y Inmigrantes en el eje Y, para que vaya avanzando
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) # es parte de Matplotlib, y se usa para personalizar las marcas (ticks) del eje X en un gráfico. Esto es útil cuando tienes muchos valores en el eje X (por ejemplo, todos los años entre 1980 y 2013), y quieres mostrar solo algunos para mayor claridad.
plt.yticks([350,1100,2500,3500,4500,5500,6500]) # es parte de Matplotlib, y se usa para personalizar las marcas (ticks) del eje X en un gráfico. Esto es útil cuando tienes muchos valores en el eje Y (por ejemplo, todos los años entre 1980 y 2013), y quieres mostrar solo algunos para mayor claridad.
plt.show()
```

```python
plt.figure(figsize=(8,4)) # Con est código se establece el tamaño de la figura
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])  #Grafico Años en el eje X y Inmigrantes en el eje Y, para que vaya avanzando
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) # es parte de Matplotlib, y se usa para personalizar las marcas (ticks) del eje X en un gráfico. Esto es útil cuando tienes muchos valores en el eje X (por ejemplo, todos los años entre 1980 y 2013), y quieres mostrar solo algunos para mayor claridad.
plt.yticks([350,1100,2500,3500,4500,5500,6500]) # es parte de Matplotlib, y se usa para personalizar las marcas (ticks) del eje X en un gráfico. Esto es útil cuando tienes muchos valores en el eje Y (por ejemplo, todos los años entre 1980 y 2013), y quieres mostrar solo algunos para mayor claridad.
plt.title('Inmigracion de Colombianos hacia Canada') # Se define un titulo para el grafico
plt.xlabel('Año') # Se define una etiqueta para el eje x
plt.ylabel('Numero Inmigrantes') #Se define una etiqueta para el eje y
plt.show() # Este debe ser el último comando
```

#### Haga lo que hicimos

#### Generar variables para graficos

```python
fig,ax = plt.subplots(figsize=(8,4)) #Crea una figura (fig) y un eje (ax) para graficar.

ax.plot(datos_col['Año'],datos_col['Inmigrantes']) #Dibuja una gráfica de línea en el eje ax.
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #ax.xaxis Se refiere al eje X del gráfico que se está dibujando en ax. Es un objeto de tipo XAxis, que puedes personalizar (frecuencia de las marcas, etiquetas, etc.).
#.set_major_locator(...) Esta función permite establecer la ubicación de las marcas principales (ticks) del eje X. Las "major ticks" son las marcas grandes del eje (por defecto, donde se muestran los números como los años).
# plt.MultipleLocator(5) Esto crea un "locator" que le indica al eje dónde colocar esas marcas principales.
ax.set_title('Inmigracion de Colombianos hacia Canada\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Numero Inmigrantes')
plt.show()
```

#### Creando Subplots en una dirección

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3)) # se creo 1 fila y 2 columnas, Crea una figura (fig) y dos subgráficos lado a lado (axs[0] y axs[1]).

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes']) #Dibuja un gráfico de líneas en el primer subplot (axs[0]) con: Eje X: los años. Eje Y: número de inmigrantes.
axs[0].set_title('Inmigración de Colombianos hacia Canada\n 1980 - 2013')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].set_xlabel('Año')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid() # coloca cuadricula en el grafico

axs[1].boxplot(datos_col['Inmigrantes']) #Grafico de cajas, muestra datos estadisticos
axs[1].set_title('Boxplot de inmigración de Colombia\nhacia Canada 1980 - 2013')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].set_xlabel('Colombia')
axs[1].grid() # coloca cuadricula en el grafico

plt.show()
```

#### Creando Subplots en dos direcciones #1

```python
fig , axs = plt.subplots(2 , 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5,wspace=0.3) # Ajusta la separación horizontal entre los subgráficos
fig.suptitle('Inmigracion de los 4 mayores paises Sudamericanos\nhacia Canada 1980 - 2013') # Se asigna un titulo a todo el grafico

axs[0,0].plot(datos.loc['Colombia', anos]) # Primer elemento de la primera fila
axs[0,0].set_title('Colombia')

axs[0,1].plot(datos.loc['Brasil', anos]) # segundo elemento de la primera fila
axs[0,1].set_title('Brasil')

axs[1,0].plot(datos.loc['Argentina', anos]) # Primer elemento de la segunda fila
axs[1,0].set_title('Argentina')

axs[1,1].plot(datos.loc['Perú', anos]) # Segundo elemento de la segunda fila
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel(): #" ".ravel()"" aplana esa matriz en una lista unidimensional de ejes, lo cual facilita iterar sobre todos los subgráficos, sin importar si estaban en filas o columnas.
  ax.set_ylim(ymin,ymax) #Establece los límites del eje Y (vertical) en el subgráfico ax. ymin y ymax deben ser valores numéricos, y representan el mínimo y el máximo visible en el eje Y.


for ax in axs.flat: #" axs" es una matriz (o lista) de subgráficos creada con plt.subplots(). , ".flat" convierte la matriz (aunque tenga solo una fila o columna) en un iterador plano.
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Año')
    ax.set_ylabel('Número de Inmigrantes')
    ax.grid()

plt.show()
```

#### Haga lo que hicimos - creando figuras con matplotlib - variables

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
df_tienda = pd.DataFrame(ventas_2022, index=tiendas) #Para facilitar la creación de los subgráficos, puedes definir la columna "tiendas" como índice del DataFrame y utilizar la propiedad loc de la biblioteca Pandas para trazar cada una de las tiendas.
df_tienda
```

```python
## Crear la figura y los subgráficos
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

#Ajustar los tamaños entre graficos
fig.subplots_adjust(hspace=0.4 , wspace=0.3)

#Agregamos un titulo general para todo el grafico
fig.suptitle('Ventas en el periodo de enero a diciembre del año 2022 en las tiendas A, B, C y D')

#Agregar los graficos en cada uno de los subgraficos

axs[0, 0].plot(df_tienda.loc['A']) # Primer elemento de la primera fila
axs[0, 0].set_title('Ventas en la Tienda A')

axs[0, 1].plot(df_tienda.loc['B']) # Segundo elemento de la primera fila
axs[0, 1].set_title('Ventas en la Tienda B')

axs[1, 0].plot(df_tienda.loc['C']) # Primer elemento de la segunda fila
axs[1, 0].set_title('Ventas en la Tienda C')

axs[1, 1].plot(df_tienda.loc['D']) #Segundo elemento de la segunda fila
axs[1, 1].set_title('Ventas en la Tienda D')

ymin = 0
ymax = 450

for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)

 # Agregar etiquetas para los ejes X y Y

for ax in axs.flat:
    ax.set_xlabel('Mes')
    ax.set_ylabel('Ventas')
    ax.grid()

plt.show()
```

#### Personalizando Graficos con Matplotlib

```python
fig,ax = plt.subplots(figsize=(8,3)) #Crea una figura (fig) y un eje (ax) para graficar.

ax.plot(datos_col['Año'],datos_col['Inmigrantes'], lw= 2, marker='o') #Dibuja una gráfica de línea en el eje ax.  lw= 3 permite aumentar el grosor de la linea, marker='o' permite colocar puntos
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #ax.xaxis Se refiere al eje X del gráfico que se está dibujando en ax. Es un objeto de tipo XAxis, que puedes personalizar (frecuencia de las marcas, etiquetas, etc.).
#.set_major_locator(...) Esta función permite establecer la ubicación de las marcas principales (ticks) del eje X. Las "major ticks" son las marcas grandes del eje (por defecto, donde se muestran los números como los años).
# plt.MultipleLocator(5) Esto crea un "locator" que le indica al eje dónde colocar esas marcas principales.
ax.set_title('Inmigracion de Colombianos hacia Canada\n1980 - 2013', fontsize= 18, loc='left') #fontsize modifica el tamaño de la fuente , loc='left' permite colocar el titulo a la izquierda
ax.set_xlabel('Año', fontsize= 14)
ax.set_ylabel('Numero Inmigrantes',fontsize= 14)
ax.yaxis.set_tick_params(labelsize= 12) # axis = eje
ax.xaxis.set_tick_params(labelsize= 12) # set_tick_params(...): es un método que te permite configurar las marcas de los ejes (los ticks), como: amaño de las etiquetas dirección,color,grosor,visibilidad
ax.grid()
plt.show()
```
