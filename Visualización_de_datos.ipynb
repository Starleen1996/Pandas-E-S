## Visualización de datos con Python: Matplotlib,Seaborn y Plotly

```python
import pandas as pd
```

```python
datos = pd.read_csv('/content/inmigrantes_canada.csv')
print(datos.head(5))
```

```python
datos.head()
```

```python
datos.shape
```

```python
datos.info() #object = string caracteres
```

```python
print(datos.columns)
```

```python
datos.columns = datos.columns.str.strip()
```

```python
print(datos.columns.tolist())
```

```python
datos.set_index('Pais', inplace=True)
datos
```

```python
#Crea una lista de strings (cadenas de texto) que van desde el año 1980 hasta el 2013 (el range en Python excluye el último número, por eso no incluye el 2014).
anos = list(map(str, range(1980, 2014)))
anos
```

```python
#.loc[] es una forma de acceder a datos por etiquetas en pandas.
# Se refiere a la etiqueta de la fila. Para que esto funcione, el índice (index) del DataFrame datos debe tener la columna 'Pais' configurada como índice, lo cual ya hiciste antes con:
colombia = datos.loc['Colombia', anos]
colombia
```

```python
#Ese fragmento de código convierte los datos de inmigración de Colombia en un diccionario Python que será útil, por ejemplo, para crear un DataFrame o visualizar los datos. Vamos paso a paso:
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4)) # Con est código se establece el tamaño de la figura
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])  #Grafico Años en el eje X y Inmigrantes en el eje Y, para que vaya avanzando
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) # es parte de Matplotlib, y se usa para personalizar las marcas (ticks) del eje X en un gráfico. Esto es útil cuando tienes muchos valores en el eje X (por ejemplo, todos los años entre 1980 y 2013), y quieres mostrar solo algunos para mayor claridad.
plt.yticks([350,1100,2500,3500,4500,5500,6500]) # es parte de Matplotlib, y se usa para personalizar las marcas (ticks) del eje X en un gráfico. Esto es útil cuando tienes muchos valores en el eje Y (por ejemplo, todos los años entre 1980 y 2013), y quieres mostrar solo algunos para mayor claridad.
plt.show()
```

```python
plt.figure(figsize=(8,4)) # Con est código se establece el tamaño de la figura
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])  #Grafico Años en el eje X y Inmigrantes en el eje Y, para que vaya avanzando
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) # es parte de Matplotlib, y se usa para personalizar las marcas (ticks) del eje X en un gráfico. Esto es útil cuando tienes muchos valores en el eje X (por ejemplo, todos los años entre 1980 y 2013), y quieres mostrar solo algunos para mayor claridad.
plt.yticks([350,1100,2500,3500,4500,5500,6500]) # es parte de Matplotlib, y se usa para personalizar las marcas (ticks) del eje X en un gráfico. Esto es útil cuando tienes muchos valores en el eje Y (por ejemplo, todos los años entre 1980 y 2013), y quieres mostrar solo algunos para mayor claridad.
plt.title('Inmigracion de Colombianos hacia Canada') # Se define un titulo para el grafico
plt.xlabel('Año') # Se define una etiqueta para el eje x
plt.ylabel('Numero Inmigrantes') #Se define una etiqueta para el eje y
plt.show() # Este debe ser el último comando
```

#### Haga lo que hicimos

```python
df_comparacion = datos.loc[['Brasil', 'Argentina'], anos]
```

```python
#El DataFrame df_comparacion se puede transponer para intercambiar las filas por las columnas y las columnas por las filas del DataFrame.
#Luego, podemos usar la propiedad .T (abreviatura de transpose):
df_comparacion = df_comparacion.T
```

```python
#Como resultado, tenemos una columna con los años y dos columnas para cada país que contienen el número de inmigrantes.
df_comparacion.head()
```

```python
plt.plot(df_comparacion['Brasil'], label='Brasil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```
