## Visualización de datos con Python: Matplotlib,Seaborn y Plotly

```python
import pandas as pd
```

```python
datos = pd.read_csv('/content/inmigrantes_canada.csv')
print(datos.head(5))
```

```python
datos.columns = datos.columns.str.strip()
```

```python
print(datos.columns.tolist())
```

```python
datos.set_index('Pais', inplace=True)
datos
```

```python
#Crea una lista de strings (cadenas de texto) que van desde el año 1980 hasta el 2013 (el range en Python excluye el último número, por eso no incluye el 2014).
anos = list(map(str, range(1980, 2014)))
anos
```

```python
#.loc[] es una forma de acceder a datos por etiquetas en pandas.
# Se refiere a la etiqueta de la fila. Para que esto funcione, el índice (index) del DataFrame datos debe tener la columna 'Pais' configurada como índice, lo cual ya hiciste antes con:
colombia = datos.loc['Colombia', anos]
colombia
```

```python
#Ese fragmento de código convierte los datos de inmigración de Colombia en un diccionario Python que será útil, por ejemplo, para crear un DataFrame o visualizar los datos. Vamos paso a paso:
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

#### Creando Subplots en dos direcciones #1

```python
fig , axs = plt.subplots(2 , 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5,wspace=0.3) # Ajusta la separación horizontal entre los subgráficos
fig.suptitle('Inmigracion de los 4 mayores paises Sudamericanos\nhacia Canada 1980 - 2013') # Se asigna un titulo a todo el grafico

axs[0,0].plot(datos.loc['Colombia', anos]) # Primer elemento de la primera fila
axs[0,0].set_title('Colombia')

axs[0,1].plot(datos.loc['Brasil', anos]) # segundo elemento de la primera fila
axs[0,1].set_title('Brasil')

axs[1,0].plot(datos.loc['Argentina', anos]) # Primer elemento de la segunda fila
axs[1,0].set_title('Argentina')

axs[1,1].plot(datos.loc['Perú', anos]) # Segundo elemento de la segunda fila
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel(): #" ".ravel()"" aplana esa matriz en una lista unidimensional de ejes, lo cual facilita iterar sobre todos los subgráficos, sin importar si estaban en filas o columnas.
  ax.set_ylim(ymin,ymax) #Establece los límites del eje Y (vertical) en el subgráfico ax. ymin y ymax deben ser valores numéricos, y representan el mínimo y el máximo visible en el eje Y.


for ax in axs.flat: #" axs" es una matriz (o lista) de subgráficos creada con plt.subplots(). , ".flat" convierte la matriz (aunque tenga solo una fila o columna) en un iterador plano.
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Año')
    ax.set_ylabel('Número de Inmigrantes')
    ax.grid()

plt.show()
```

#### Haga lo que hicimos - creando figuras con matplotlib - variables

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
df_tienda = pd.DataFrame(ventas_2022, index=tiendas) #Para facilitar la creación de los subgráficos, puedes definir la columna "tiendas" como índice del DataFrame y utilizar la propiedad loc de la biblioteca Pandas para trazar cada una de las tiendas.
df_tienda
```

```python
## Crear la figura y los subgráficos
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

#Ajustar los tamaños entre graficos
fig.subplots_adjust(hspace=0.4 , wspace=0.3)

#Agregamos un titulo general para todo el grafico
fig.suptitle('Ventas en el periodo de enero a diciembre del año 2022 en las tiendas A, B, C y D')

#Agregar los graficos en cada uno de los subgraficos

axs[0, 0].plot(df_tienda.loc['A']) # Primer elemento de la primera fila
axs[0, 0].set_title('Ventas en la Tienda A')

axs[0, 1].plot(df_tienda.loc['B']) # Segundo elemento de la primera fila
axs[0, 1].set_title('Ventas en la Tienda B')

axs[1, 0].plot(df_tienda.loc['C']) # Primer elemento de la segunda fila
axs[1, 0].set_title('Ventas en la Tienda C')

axs[1, 1].plot(df_tienda.loc['D']) #Segundo elemento de la segunda fila
axs[1, 1].set_title('Ventas en la Tienda D')

ymin = 0
ymax = 450

for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)

 # Agregar etiquetas para los ejes X y Y

for ax in axs.flat:
    ax.set_xlabel('Mes')
    ax.set_ylabel('Ventas')
    ax.grid()

plt.show()
```

#### Haga lo que hicimos - parte 2

1. Cambiar la posición de los títulos de los subgráficos a la izquierda.
2. Aumentar el tamaño de la fuente del título general de la figura para resaltarlo.
3. Aumentar el tamaño de los títulos y etiquetas de los ejes de los subgráficos.
4. Hacer que las líneas sean más gruesas.
Cambiar el color de las líneas de cada tienda para diferenciarlas aún más.

```python
## Crear la figura y los subgráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 6))

#Ajustar los tamaños entre graficos
fig.subplots_adjust(hspace=0.4 , wspace=0.3)

#Agregamos un titulo general para todo el grafico
fig.suptitle('Ventas en el periodo de enero a diciembre del año 2022 en las tiendas A, B, C y D', fontsize=20)

#Generar una lista par incluir los colores del grafico
colores =['darkviolet', 'green', 'darkblue', 'coral']

#Generar bucle para trazar y generar graficos

for i, ax in enumerate(axs.flat):
    ax.plot(df_tienda.columns, df_tienda.iloc[i], color=colores[i], lw=3) #Dibuja una gráfica de línea en el eje ax.  lw= 3 permite aumentar el grosor de la linea, marker='o' permite colocar puntos
    # df_tienda.iloc[i]: accede a la fila i del DataFrame. df_tienda.columns: se usa como eje X, es decir, los meses. color=colores[i]: color del gráfico. lw=3: grosor de la línea.
    ax.set_title(f'Ventas en la tienda {df_tienda.index[i]}', loc='left', fontsize=16) # Realiza una iteración en cada tienda para establecer el titulo, en la izquierda con un tamaño de fuente de 16
    ax.set_xlabel('Mes', fontsize=14) # Define el tamaño de la fuente en 14
    ax.set_ylabel('Número de Ventas', fontsize=14)  # Define el tamaño de la fuente en 14
    ax.tick_params(labelsize=12)
    ax.grid(linestyle='--',color='lightgrey')

plt.show()
```

```python
# Vamos a definir una función para probar diferentes paletas de colores

def generar_grafico(palette):
  fig , ax =plt.subplots(figsize=(8,4))

  ax.set_frame_on(False) #Oculta el marco o borde exterior del gráfico (también llamado "spines" en Matplotlib).
  #ax.xaxis.set_visible(False) #Oculta completamente el eje X, El eje X desaparece del gráfico.
  #ax.tick_params(axis='y',which='both',size=0) #Modifica los parámetros visuales de los ticks (marcas del eje) en el eje Y. Específicamente: axis='y' Afecta solo al eje vertical (Y). which='both' Los ticks mayores (los principales, normalmente más largos).Los ticks menores (si existieran).size=0
  # Hace que el tamaño de las líneas de los ticks sea cero, es decir, invisibles.

  ax = sns.barplot(data=top_10, x='Total',y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 Paises con mayor migración hacia Canada\n1981 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  for i,j in enumerate(top_10['Total']):  # Permite realizar anotaciones o colocar etiquetas en cada una de las barras
   ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```

```python
# Vamos a definir una función para probar diferentes paletas de colores

def generar_grafico(palette):
  fig , ax =plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[]) #xticklabels=[]: le está diciendo a Matplotlib que las etiquetas del eje X deben ser una lista vacía, es decir, que no se muestren.
  ax = sns.barplot(data=top_10, x='Total',y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 Paises con mayor migración hacia Canada\n1981 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  sns.despine() #Elimina los bordes de los graficos
  for i,j in enumerate(top_10['Total']):  # Permite realizar anotaciones o colocar etiquetas en cada una de las barras
   ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```
